## 입도객 데이터

```python
Data_3 = pd.read_excel('../data/train_data/제주도내외국인입도통계.xlsx')
Data_3
```

```python
Data_3 = Data_3.sort_values(by='base_date')
```

```python
size = [31,28,31,30,31,30,31,31,30,31,30,31,31,28,31,30,31,30,31,31,30,31,30,31,31,29,31,30,31,30,31,31,30,31,30,31,31,28,31,30,31,30]
```

```python
Data_3['일평균내국인입도'] = Data_3['내국인입도현황']/size
```

```python
Data_3['일평균외국인입도'] = Data_3['외국인입도현황']/size
```

```python
Data_3.drop(['내국인입도현황','외국인입도현황'],axis=1,inplace=True)
```

```python
Data_3['일평균내국인입도'] = Data_3['일평균내국인입도'].astype('int')
Data_3['일평균외국인입도'] = Data_3['일평균외국인입도'].astype('int')
```

```python
Date = pd.date_range('2018-01-01','2021-06-30',freq="D")
Date
```

```python
new_Date = []
for i in range(len(Date)):
    new_Date.append(str(Date[i])[:10])
```

```python
a = pd.DataFrame(columns=['일자','월'])
```

```python
a['일자'] = new_Date
```

```python
month = []
for i in range(len(a)):
    month.append(a['일자'][i][:7])
```

```python
a['월'] = month
```

```python
a['월'] = Data_3['base_date']
```

```python
Data_3['일자'] = Data_3['base_date']
```

```python
Data_3.drop('base_date',axis=1,inplace=True)
```

```python
Data_3
```

```python
a.columns = ['임시','일자']
```

```python
b = pd.merge(a, Data_3, on='일자')
b.drop('일자',axis=1,inplace=True)
b.columns = ['일자','일평균내국인입도','일평균외국인입도']
```

```python
b.to_csv('../data/processed_data/일평균입도객평균.csv',encoding='cp949',index=False)
```
